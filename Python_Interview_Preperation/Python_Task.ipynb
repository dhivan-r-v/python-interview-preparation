## Rock, Paper, Scissors Game

```python

import random
computer = ["rock","paper","scissors"]
while True:
    computer_choice=random.choice(computer)
    player=input("Enter the choice:")
    player_choice=player.lower()
    if player_choice not in computer:
        print("Invalid Input!Enter correct input")
        continue
    if player_choice==computer_choice:
        print("Draw")
    elif (player_choice=="rock" and computer_choice=="scissors")or(player_choice=="paper" and computer_choice=="rock")or(player_choice=="scissors" and computer_choice=="paper"):
        print("Computer Choice:{},Player Choice:{},You win".format(computer_choice,player_choice))
    else:
        print("Computer Choice:{},Player Choice:{},Computer Wins".format(computer_choice,player_choice))
    

    while True:
        playagain=input("Enter Yes/No:").lower()
        if playagain!="yes" and playagain!="no":
            print("Enter only Yes/No")
            continue
        break
        
    if playagain=="yes":
        print("Let's Continue the game")
        continue
    elif playagain=="no":
        print("Let's end the game")
        break

    
```

### Rock,Paper ,Scissor using user defined function

```python
import random
choices=["rock","paper","scissors"]
def playerchoice(choices):
    while True:
        player_choice=input("Enter rock/paper/scissors:").lower()
        if player_choice not in choices:
            print("Invalid input,Enter correct input")
            continue
        return player_choice
def gameplay(choices):
    computer_choice=random.choice(choices)
    human_choice=playerchoice(choices)
    if computer_choice==human_choice:
        print("Game Draw")
    elif (human_choice=="rock" and computer_choice=="scissors")or(human_choice=="paper" and computer_choice=="rock")or(human_choice=="scissors" and computer_choice=="paper"):
        print("You won , The computer Choice is {} and the Human Choice is {}".format(computer_choice,human_choice))
    else:
        print("Computer Won , The Computer choice is {} and the Human Choice is {}".format(computer_choice,human_choice))
def playagain():
    while True:
        again=input("Enter Yes/No:").lower()
        if again in ["yes","no"]:
            return again
        print("Enter yes or no only:")

while True:
    game_play=gameplay(choices)
    play_again=playagain()
    if play_again=="yes":
        print("Let's continue the game")
    else:
        print("Let's end the game")
        break    
    
```

```python
import random
words=["society","glassceiling","dyarchy","ascribed","meritocracy","monarch"]

def wordgame(words):
    randomword=random.choice(words)
    blank=["_"]*len(randomword)
    return randomword,blank

def gamestarts(randomword,blank):
    attempts=6
    guessed_letter=[]
    while attempts>0:
        print("\n Word:"," ".join(blank))
        print("Attempts Left",attempts)
        guess=input("Enter the guessed letter:").lower()
        if len(guess)!=1 or not guess.isalpha():
            print("Enter a Single valid letter!")
            continue
        if guess in guessed_letter:
            print("You guessed the letter")
            continue
        guessed_letter.append(guess)

        if guess in randomword:
            print("Correct Guess")

            for i in range(len(randomword)):
                if randomword[i]==guess:
                    blank[i]=guess
        else:
            print("Wrong Guess")
            attempts -=1

        if "_" not in blank:
            print("You Win. The word is",randomword)
            return
    print("\n You Lose. The word is",randomword)

randomword, blank = wordgame(words)
gamestarts(randomword, blank)

```
