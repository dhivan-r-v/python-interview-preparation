## Sets

### add

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
developed_countries.add("Russia")
print(developed_countries)
```

### clear

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
developed_countries.clear()
print(developed_countries)
```

### copy

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
countries=developed_countries.copy()
print(countries)
```

### difference

```python
## difference removes common element and returns remaining element in developed_countries

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.difference(high_gdp_countries)
```

### difference_update

```python
## same like difference but here it updates to the original list

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.difference_update(high_gdp_countries)
print(developed_countries)
```

### discard

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
countries=developed_countries.discard("Finland")
print(developed_countries)
```

### intersection

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
countries=developed_countries.intersection(high_gdp_countries)
print(countries)
```

### intersection_update

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.intersection_update(developed_countries)
print(developed_countries)
```

### isdisjoint

```python
##isdisjoint: check whether no common elements are present.

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.isdisjoint(high_gdp_countries)
```

### issubset 

```python
##issubset: check whether all the elements of setA are present in setB.

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.issubset(high_gdp_countries)
```

### issuperset

```python
## issuperset : checks whether all elements of a contains b

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.issuperset(high_gdp_countries)
```

### pop

```python
## here pop removes any random element

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.pop()
```

### remove

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
developed_countries.remove("Finland")
developed_countries
```

### symmetric_difference

```python
##symmetric_difference : returns elements from either sets but not in both

developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.symmetric_difference(high_gdp_countries)
```

### union

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
high_gdp_countries={"India","Japan","Germany","USA",}
developed_countries.union(high_gdp_countries)
```

### update

```python
developed_countries={"Finland","Germany","Australia","France","USA"}
developed_countries.update(["Scotland"])
print(developed_countries)
```

## Dictionary

### clear

### fromkeys & Update

```python
least_developed_countries={"southsudan":3,'niger':0.5,'nigeria':3.5,'pakistan':3,'cuba':4}
new=least_developed_countries.fromkeys(["Iran","Iraq"],4)
least_developed_countries.update(new)
print(least_developed_countries)
```
